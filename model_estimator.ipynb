```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
df = pd.read_csv('eda_cleased_data.csv')
df.head()
```

```python
df.columns
```

```python
# Pulling only required columns
df_req = df[['Rating','Size','Type of ownership', 'Industry', 'Sector', 'Revenue','hourly', 'emp_provid_sal','job_location', 'same_location','age', 'python',
       'spark', 'sql', 'aws', 'job_simp', 'seniority','desc_len', 'num_comp','avg_salary']]
df_req
```

```python
# converting categorical features into numerical equivalent using One Hot encoding
df_model = pd.get_dummies(df_req)
df_model.head()
```

```python
X = df_model.drop('avg_salary', axis =1)
y = df_model.avg_salary.values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state=0)
```

# multiple linear regression 

```python
import statsmodels.api as sm

X_sm = X = sm.add_constant(X)
model = sm.OLS(y,X_sm)
model.fit().summary()
```

```python
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.model_selection import cross_val_score
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
np.mean(cross_val_score(lm,X_train,y_train, scoring = 'neg_mean_absolute_error', cv= 3))
```

```python
from sklearn.linear_model import  Lasso
```

```python
lm_l = Lasso()
lm_l.fit(X_train,y_train)
np.mean(cross_val_score(lm_l,X_train,y_train, scoring = 'neg_mean_absolute_error', cv= 3))
```

```python
alpha = []
error = []

for i in range(1,100):
    alpha.append(i/100)
    lml = Lasso(alpha=(i/100))
    error.append(np.mean(cross_val_score(lml,X_train,y_train, scoring = 'neg_mean_absolute_error', cv= 3)))
    
plt.plot(alpha,error)
```

```python
err = tuple(zip(alpha,error))
df_err = pd.DataFrame(err, columns = ['alpha','error'])
df_err[df_err.error == max(df_err.error)]
```

```python
lm_l = Lasso(alpha=0.11)
lm_l.fit(X_train,y_train)
np.mean(cross_val_score(lm_l,X_train,y_train, scoring = 'neg_mean_absolute_error', cv= 3))
```

```python
# random forest 
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
np.mean(cross_val_score(rf,X_train,y_train,scoring = 'neg_mean_absolute_error', cv= 3))
```

```python
# tune models GridsearchCV 
from sklearn.model_selection import GridSearchCV
```

```python
#GridSearchCV Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 500, num = 5)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt','log2']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 50, num = 5)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 20]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
# Method of selecting samples for training each tree
# bootstrap = [True, False]
criterion = ['mse','mae']
```

```python

# Create the GridSearchCV
grid_params = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
              'criterion' :criterion}

print(grid_params)
```

```python
# Use the GridSearchCV to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
rf_random = GridSearchCV(estimator = rf, param_grid=grid_params,scoring='neg_mean_squared_error', cv = 3, verbose=2, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
print(rf_random.best_score_)
print(rf_random.best_estimator_)
```

```python
# test ensembles 
tpred_lm = lm.predict(X_test)
tpred_lml = lm_l.predict(X_test)
tpred_rf = rf_random.best_estimator_.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,tpred_lm))
print(mean_absolute_error(y_test,tpred_lml))
print(mean_absolute_error(y_test,tpred_rf))
```

```python
mean_absolute_error(y_test,(tpred_lm+tpred_rf)/2)
```
